```python
%%sql
select*
from flipkart_table
```

# Find the best rated Laptops

```python
%%sql 
SELECT *
FROM flipkart_table
WHERE stars >= 4.5 AND Ratings >= 1500
ORDER BY stars DESC;
```

# So which Laptop was the best seller?

```python
%%sql
SELECT *
FROM flipkart_table
ORDER BY Ratings DESC
LIMIT 1;
```

# Having highest discount 

```python
%%sql 
SELECT DISTINCT ProductName
FROM flipkart_table
where MRP-CurrentPrice>30000


```
